```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Original data
data = {
    'name': ['mohan', 'maria', 'sakib', 'tao', 'virat', 'khusbu', 'dmitry', 'selena', 'john', 'imran',
             'jose', 'deepika', 'yoseph', 'binod', 'gulshan', 'johnson', 'donald', 'aamir', 'ken', 'Liu'],
    'height': [1.2, 2.3, 4.9, 5.1, 5.2, 5.4, 5.5, 5.5, 5.6, 5.6,
               5.8, 5.9, 6.0, 6.1, 6.2, 6.5, 7.1, 14.5, 23.2, 40.2]
}

df = pd.DataFrame(data)

# Calculate IQR
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outliers
df_no_outlier = df[(df.height > lower_limit) & (df.height < upper_limit)]

# Plot boxplots
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot with outliers
sns.boxplot(data=df, y='height', ax=axes[0])
axes[0].set_title('With Outliers')

# Boxplot without outliers
sns.boxplot(data=df_no_outlier, y='height', ax=axes[1])
axes[1].set_title('Without Outliers')

plt.tight_layout()
plt.show()
```
