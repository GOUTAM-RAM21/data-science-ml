```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(Dense(1 ,activation='sigmoid'))

model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=32,epochs=100,validation_data=(X_test,y_test))
```

```python
import kerastuner as kt
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(32,activation='relu',input_dim=8))
  model.add(Dense(1,activation='sigmoid'))
  optimizer=hp.Choice('optimizer',values=['adam','sgd','rmsprop','adaelta'])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5
)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
def build_model(hp):
  model = Sequential()

  # Tune the number of hidden layers
  num_layers = hp.Int('num_layers', min_value=1, max_value=10)

  # First hidden layer (always present)
  # input_dim should only be specified for the first layer
  model.add(Dense(
      hp.Int('units_0', min_value=8, max_value=128, step=8),
      activation=hp.Choice('activation_0', values=['relu', 'tanh', 'sigmoid']),
      input_dim=8
  ))

  # Additional hidden layers
  for i in range(1, num_layers): # Start from 1 because the first layer is already added
    model.add(Dense(
        hp.Int('units_' + str(i), min_value=8, max_value=128, step=8),
        activation=hp.Choice('activation_' + str(i), values=['relu', 'tanh', 'sigmoid'])
    ))

  # Output layer (outside the loop)
  model.add(Dense(1, activation='sigmoid'))

  # Compile the model (outside the loop)
  model.compile(
      optimizer=hp.Choice('optimizer', values=['adam', 'sgd', 'rmsprop']),
      loss='binary_crossentropy',
      metrics=['accuracy']
  )
  return model
```

```python
tuner_final = kt.RandomSearch(build_model, objective='val_accuracy', max_trials=10, project_name='full_model_tuning', overwrite=True)
```

```python
tuner_final.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```
